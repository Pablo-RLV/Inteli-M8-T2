## Importação de bibliotecas

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio, display
import ipywidgets as widgets

preload_models()
```

## Execução da solução

É importante salientar que o sistema apresenta dificuldades de interpretação em relação a caracteres especiais, como acentos e "ç". Portanto, a utilização desses pode trazer problemas de usabilidade.

```python
# @title Insira um texto e aperte o botão para criar um novo áudio {display-mode:"form" }

def generate_audio_button(sender):
    text_prompt = "Oi! Eu sou o Pablo! Eu sou brasileiro."  # @param {type:"string"}
    audio_array = generate_audio(text_prompt)
    display(Audio(audio_array, rate=SAMPLE_RATE))

button = widgets.Button(description="Gerar áudio")
button.on_click(generate_audio_button)
display(button)
```

### ALTERNATIVA: Tradução de português para inglês intregrada

Para essa implementação, foi utilizada a biblioteca deep-translator, que permite a tradução de textos para outros idiomas, de uma forma simplificada.

```python
!pip install deep-translator --quiet
```

```python
from deep_translator import GoogleTranslator
trans = GoogleTranslator(source_language="pt", target_language="en")

# @title Insira um texto e aperte o botão para criar um novo áudio {display-mode:"form" }

def generate_audio_button(sender):
    text_prompt = "Oi! Eu sou o Pablo! Eu sou brasileiro."  # @param {type:"string"}
    translated_text = trans.translate(text_prompt)
    print(translated_text)
    audio_array = generate_audio(translated_text)
    display(Audio(audio_array, rate=SAMPLE_RATE))

button = widgets.Button(description="Gerar áudio")
button.on_click(generate_audio_button)
display(button)

```
